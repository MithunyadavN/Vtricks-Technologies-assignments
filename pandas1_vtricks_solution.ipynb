```python
# combine all the files - market, cust_dimen , order_dimen, product_dimen , shippin dimen and name it as master dataframe

market_df=pd.read_csv("../market_fact.csv")
custDimen_df=pd.read_csv("../cust_dimen.csv")
prodDimen_df=pd.read_csv("../prod_dimen.csv")
ordersDimen_df=pd.read_csv("../orders_dimen.csv")
shippingDimen_df=pd.read_csv("../shipping_dimen.csv")
```

```python
# From now on consider only master table for data slicing
df1=pd.merge(market_df,custDimen_df,how="inner",on="Cust_id")
df2=pd.merge(df1,prodDimen_df,how="inner",on="Prod_id")
df3=pd.merge(df2,ordersDimen_df,how="inner",on="Ord_id")
master_df=pd.merge(df3,shippingDimen_df,how="inner",on="Ship_id")
print(master_df.shape)
```

```python
# how many low, medium and high priority customers are there  ?
print( master_df.groupby(["Order_Priority"])["Order_Priority"].count())
```

```python
# which kind of delivery ussually customers prefer : give your analysis by seeing the count of delivery type
print(master_df.groupby(["Ship_Mode"])["Ship_Mode"].count())

# analysi s : air delivery is the one people prefer more - analysing it might be for the fast delivery
# we can also look for delivery time and shipping cost for further analysis
```

```python
# how many customers has purchased product1 in 'west' region?
print(master_df.loc[(master_df['Prod_id']=="Prod_1") & (master_df["Region"]=="WEST")]["Cust_id"].count())
```

```python
# which customer segment are more in "nunavat" region?
custsegment=master_df[(master_df['Region']=='NUNAVUT')].groupby(["Customer_Segment"])['Customer_Segment'].count()
print("Highest Customer Segment in Nunavat is  ",custsegment.idxmax()," and count is ",custsegment.max())
```

```python
# list all the customers whose name starts with letter 'J'
print("Customer Name starting  with J are ",master_df.loc[(master_df['Customer_Name'].str.startswith('J'))]['Customer_Name'])
```

```python
# CREATE A DATAFRAME WITH PRODUCTID AND TOTAL SALES
ProductSales=master_df.groupby(["Prod_id"])["Sales"].sum()
ProductSales
```

```python
# CREATE A DATAFRAME WITH PRODUCTID AND TOTAL PROFIT

ProductProfit=master_df.groupby(["Prod_id"])["Profit"].sum()
ProductProfit
```
